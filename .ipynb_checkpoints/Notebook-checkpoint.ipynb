# Guitar Sustain Data Analysis
This is an analysis of the data that Jim Lill recorded and used as part of his [video on guitar sustain](https://youtu.be/muVzwbkUUnM) and made available on his [website](https://jimlillmusic.com/sustain/) in pdf and xlsx formats. He also provided details on the recording process.

We will examine his data and see if we can identify any major trends left unidentified, though I imagine we'll come to the same conclusions that Jim did in his video.

```python
import pandas as pd
from os.path import exists
import requests
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

file = 'Jims-sustain-test-list-and-numbers.xlsx'

if not exists(file): # Download the file if it isn't already in the notebook's directory
    r = requests.get('https://jimlillmusic.com/wp-content/uploads/2022/01/Jims-sustain-test-list-and-numbers.xlsx')
    with open(file, 'wb') as f:
        f.write(r.content)

# Create data frame        
df = pd.read_excel(file, index_col=0, header=1)
df.head(10)
```

# What We're Keeping
We want the raw recorded dB levels at 0 and 4 seconds. The original dataset tries to measure sustain, but accidentally-high peaks at the start of the sound clips might skew the numbers without indicating meaningful sustain. We should probably define what meaningful sustain is.
## Defining Sustain
In the context of guitar sustain, sustain is a note played on guitar with high enough perceived volume to be audible and present in a produced performance of music.
## What We're Doing With The Data We're Keeping
While it would be better to have the original sound files and simply measure [LUFS](https://www.sweetwater.com/insync/what-is-lufs-and-why-should-i-care/) for the first 4 seconds, we don't have that data. What we can do is convert the dB values to an amplitude value between 0 and 1 and interpolate between the two points using a simple exponential decay model: $ x(t) = a * exp(−bt) $—aka $ x(t) = a \times e^{-bt} $. After that, we'll calculate the Root Means Square or [RMS](https://mathwords.com/r/root_mean_square.htm) of the sound for the given 4 second interval to quantify sustain (Note that we will not be using [standard RMS](https://emastered.com/blog/rms-level-for-mastering) readings used in DAWs, since that is measured over a ~300ms interval). This will help minimize the impact of any wonky too-high peaks or any other abnormalities in the sound, though it won't be as good as LUFS which takes into account perceived loudness more holistically instead of approximating it like RMS. Then we'll look for trends.
          
We know that $ a $ will be the initial dB value converted to amplitude, and we will simply plug in $ t = 4 $ and solve for $ b $ based on what we know the value of $ x(4) $ is. $ b $ Should stay the same for a given setup: guitar, amp settings, etc. I imagine that, due to the inaccuracy of our model, $ b $ might vary with values of $ a $. We will compare the values of $ b $ and how they differ, but the primary purpose of this modeling will be using the model to determine an RMS level for a given pair of data points.

For reference, here is what our mathematical function might look like representing the amplitude of our guitar string over time.

```python
t = np.linspace(0, 4, num=100)
x = np.exp(-t)

plt.xlabel('time')
plt.ylabel('amplitude')
plt.plot(t, x)

plt.show()
```

# Data Cleaning

## Dropping Columns

```python
# Display indexes and column names
print(df.index, '\n'*2, df.columns)
```

```python
# Drop columns that are all NaN
df.dropna(axis=1, how='all')

# Start new variable that we will filter
raw_columns = list(df.columns)

# Cutoff columns after the last column name that we want
cutoff = raw_columns.index('High E dist hi 4sec') + 1
raw_columns = raw_columns[:cutoff]

# Filter out columns that calculate a difference
raw_columns = (col for col in raw_columns if 'diff' not in col)

df = df[raw_columns]

df.columns
```

```python
# We will set a threshold of 2 because we want to keep all lines with atleast 2 values
# Otherwise we will inadvertently keep the row that has a comment noting to double check some data.
threshold = 2

# Because of how we cleaned the columns, we can drop all of the rows that we want to drop with this line:
df.dropna(axis=0, thresh=threshold, inplace=True) 

print(df.index, '\n'*2, df.columns)
```

## Renaming Columns
We will use our indices to represent guitar, playing position, and playing intensity.

`example: strat_soft, gretsch_hard, tele_neck, tele_bridge, sg_direct_input, sg_mic`

We will use our columns to represent beginning volume and ending volume, full strum or individual note in scientific pitch notation, and amp setup.

`example: strum_0, strum_4 e2_0, e2_4, e4_low_distortion_0, e4_low_distortion_4, strum_high_distortion_0, strum_high_distortion_4.`

```python
new_column_names = iter((
    'strum_0', 'strum_4',
    'e2_0', 'e2_4',
    'e4_0', 'e4_4',
    
    'strum_low_distortion_0', 'strum_low_distortion_4',
    'e2_low_distortion_0', 'e2_low_distortion_4',
    'e4_low_distortion_0', 'e4_low_distortion_4',
    
    'strum_high_distortion_0', 'strum_high_distortion_4',
    'e2_high_distortion_0', 'e2_high_distortion_4',
    'e4_high_distortion_0', 'e4_high_distortion_4'
))
new_index_names = iter((
    'tele_direct_input', 'sg_direct_input', 'gretsch_direct_input', 'strat_direct_input',
    'tele_mic', 'sg_mic', 'gretsch_mic', 'strat_mic',
    'tele_hard', 'sg_hard', 'gretsch_hard', 'strat_hard',
    'tele_soft', 'sg_soft', 'gretsch_soft', 'strat_soft',
    'tele_bridge', 'sg_bridge', 'gretsch_bridge', 'strat_bridge',
    'tele_neck', 'sg_neck', 'gretsch_neck', 'strat_neck',
    'tele_finger', 'sg_finger', 'gretsch_finger', 'strat_finger'
))

column_dict = {original: new for original, new in zip(df.columns, new_column_names)}
index_dict = {original: new for original, new in zip(df.index, new_index_names)}

df = df.rename(columns=column_dict)
df = df.rename(index=index_dict)


df
```
